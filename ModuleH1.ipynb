**Chapter 22 - ODE Initial Value Problems**

*22.1, 22.2, & 22.3: ODE Initial Value Problem Statement, Reduction of Order, and The Euler Method*

Ordinary differential equations (ODE's) are models representing the relationship between an independent variable (t) and an arbitrary number of derivatives within the same dimension (e.g. $ R^2$). For ODE's with more than one derivative, the ODE must be re-written as that of a first derivative.


The Euler Method is an iterative technique used to approximate the solution to an ODE based on its initial conditions. Put simply, the Euler Method considers the initial condition, estimates the next point in the solution, then uses this estimation to approximate the next function value. This method will repeat these proedures for as many steps desired. For larger step sizes, the accuracy of Euler's method drops as the independent variable approaches infinity. For smaller step sizes, the accuracy of Euler's method is much higher given there are smaller breaks between estimations and thus more points to consider (and less to miss).

Below is a visual demonstration of the Euler Method.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
f = lambda t, s: t**2                                 # simple ODE function 
h1 = 0.1                                              # 0.1 step size
h2 = 0.01                                             # 0.01 step size
t = np.arange(0, 1 + h1, h1)                          # t values based on 0.1 step size
t = np.arange(0, 1 + h2, h2)                          # t values based on 0.01 step size
s0 = 1                                                # initial condition (0, 1)
exact = (1/3) * t**3 + 1                              # exact solution to f

# ODE Approximation Step Size 0.1
s_h1 = np.zeros(len(t))                               # empty array for storing results for 0.1 step size
s_h2 = np.zeros(len(t))                               # empty array for 0.01 step size
s_h1[0] = s0                                          # set initial condition for 0.1 step size
s_h2[0] = s0                                          # initial condition for 0.01 step size

for i in range(0, len(t) - 1):                        # for every point except the initial condition
    s_h1[i + 1] = s_h1[i] + h1*f(t[i], s_h1[i])       # explicit Euler method, 0.1 step size

for i in range(0, len(t) - 1):                        # for every point except the initial condition
    s_h2[i + 1] = s_h2[i] + h2*f(t[i], s_h2[i])       # explicit Euler method, 0.01 step size

# Plotting ODE Approximation vs Exact Solution 
plt.figure();
plt.plot(t, s_h1, 'b.', label='Approximate, h = 0.1');
plt.plot(t, s_h2, 'g.', label='Approximate, h = 0.01');
plt.plot(t, exact, 'k', label='Exact');
plt.title('Comparing Accuracy of Euler Approximations with Varying Step Sizes');
plt.xlabel('t');
plt.ylabel('f(t)');
plt.grid();
plt.legend(loc='upper left');
plt.show()
```
