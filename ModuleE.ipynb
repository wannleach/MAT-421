**3.2: Continuity and Differentiation**

*3.2.1: Limits and Continuity*

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# finding the limit using sympy
x_var = sp.Symbol('x')              # variable x
y_func = x_var**2                   # function y
limit = sp.limit(y_func, x_var, 2)  # the limit as function 'y' goes to '2' with respect to 'x'
print("Limit x-->2: " + str(limit))

# verifying by plotting
x = np.linspace(-5,5,100)
y = x**2
plt.figure();
plt.plot(x, y, 'k');
plt.plot(2, limit, "o", label="limit");
plt.legend();
plt.show();
```

*3.2.2: Derivatives*

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

#defining original function
x_var = sp.Symbol('x')                      # variable x
y_func = x_var**3 + x_var**2 + x_var        # function y

#finding derivatives with respect to x
first_dx = sp.diff(y_func, x_var)
second_dx = sp.diff(first_dx, x_var)
print("Original function: " + str(y_func))
print("First derivative:" + str(first_dx))
print("Second derivative:" + str(second_dx) + "\n")

#plotting for comparison
x = np.linspace(-5,5,100)                   # x-values
y0 = x**3 + x**2 + x                        # original function
y1 = 3*x**2 + 2*x + 1                       # first derivative
y2 = 6*x + 2                                # second derivative
plt.figure();
plt.plot(x, y0, 'k', label='original function');
plt.plot(x, y1, 'b', label='first derivative')
plt.plot(x, y2, 'r', label='second derivative')
plt.legend();
```

*3.2.3: Taylor's Theorem*

```python
import numpy as np
import matplotlib.pyplot as plt

# compute grid for function
h = 0.1 # step size
x = np.linspace(-5,5,100) # x-values
y = x**3 + x**2 + x # function

# first deriv finite difference approx with Taylor Series
forward_diff = np.diff(y)/h # compute vector of forward differences
x_diff = x[:-1:] # compute corresponding grid
exact_solution = 3*x_diff**2 + 2*x_diff + 1 # compute exact solution

# plotting
plt.figure();
plt.plot(x_diff, forward_diff, '--b', label = 'finite difference approximation')
plt.plot(x_diff, exact_solution, 'r', label = 'exact derivative solution');
plt.legend();
plt.show();

# Compute max error
max_error = max(abs(exact_solution - forward_diff))
print("Max error: " + str(max_error))
```

**3.3: Unconstrained Optimization**

*3.3.1: Necessary and Sufficient Conditions of Local Minimizers*

```python
from scipy import optimize
import matplotlib.pyplot as plt

# function
x = np.linspace(-5,5,100) # x-values
def y(x):                 # function
  return x**2 + 5*np.cos(x)   
grid = (-10, 10, 0.1)

# local minimum (minimum within a certain domain)
xmin_local = optimize.fminbound(y, 0, 10)
print("Local minimum found %s" % xmin_local)

# global minima (overall mimimum for function)
xmin_global = optimize.brute(y, (grid, ))
print("Global minima found %s" % xmin_global)

# plotting for comparison
plt.figure();
plt.plot(x, y(x));
```
